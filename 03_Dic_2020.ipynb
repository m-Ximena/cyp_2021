```
#Ejemplo 3 
email=False
for i in "ariel@unam.mx":
  if(i=="@"):
    email=True

if email==True:
  print("El email es correcto")

else:
  print("El email no es correcto")
```

```
email=False
miemail=input("Introduce tu correo electronico ")

for i in miemail:
  if(i=="@"):
    email=True

if email:
  print("El email es correcto")

else:
  print("El email no es correcto")
```
